```python
from gensim.models import word2vec
from gensim import models

model = models.Word2Vec.load('word2vec.model')
ex=emotionless.copy()

for i in emotionless.keys(): 
    print (len(emotionless[i]))
    for j in emotionless[i]:
        try:
            res=model.most_similar_cosmul(j, topn=10)
            
            for item in res:     
                a=item[0]
                b=item[1]
                
                if len(emotionless[i])<=2000:
                    if b > 0.9:
                        if a not in emotionless[i]:
                            emotionless[i].append(str(a))
                            print(i+"新增"+a+" "+str(b))
                            print (len(emotionless[i]))
        except Exception as e: 
            print(repr(e))
    print(i+"end")
```
