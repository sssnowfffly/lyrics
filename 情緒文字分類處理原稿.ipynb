```python
from gensim.models import word2vec
from gensim import models
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = models.Word2Vec.load('word2vec.model')
while True:
    query = open("input.txt",'r',encoding = 'utf8')
    q_list = query.read().split(" ")
    dictwordkey={}

    for i in range(len(q_list)):
        print(len(q_list)-i)
        try:
            simword=[]
            res = model.most_similar_cosmul(q_list[i],topn = 1000)
                    
            for item in res:     
                a=item[0]
                simword.append(str(a))
                    
            dictwordkey[q_list[i]]=simword
                    
        except Exception as e:
            simword.append(repr(e))
            continue
                
            print("----------------------------")

    break
    f.close()
    query.close()
```

```python
countval=[]
countword=[]

for keyword in dictwordkey.keys():      
    countnum=[]
    for inword in dictwordkey[keyword]:    
        for i in range(-1,20):
            a=str(itemize[i+1])
            if inword in datain[a]:
                countnum.append(a)
    if len(countnum)!=0:
        countword.append(keyword)
        countval.append(countnum)
```

```python
worddict={}
wordemotcount={}

for j in range(len(countval)):
    for i in itemize:
        a=countval[j].count(str(i))
        wordemotcount[str(i)]=a

    b=sorted(wordemotcount.items(),key=lambda item:item[1], reverse = 1 )[0][0]

    worddict[str(countword[j])]=b
```
