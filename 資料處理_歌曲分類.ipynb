```python
singer=[]
with open ("singer.txt","r",encoding = 'utf8') as f:
    singer=f.read().split("\n")
singer
```

```python
from bs4 import BeautifulSoup
import requests

links=[]
for sing in range(len(singer)):
    url=("https://mojim.com/"+"{}"+".html").format(singer[sing])
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html5lib')
    ll=soup.select("span.mxsh_ss3 > a[href]")
    
    for i in ll:
        a=str(i).split('<a href="')[1].split('" title="歌詞')[0]
        sr=str("https://mojim.com"+a)
        links.append(sr)
    
    print(str(singer[sing])+" song's link end")
```

```python
##語庫資料鍵入
emotionless={}
with open('emotionless1.txt', 'r',encoding = 'utf8') as f:
    a=f.read().split("\t")
    for i in range(0,len(a),2):
        label=a[i]
        inword=a[i+1].split(" ")
        emotionless[label]=inword
        
itemize=["NA","NB","NC","ND","NE","NG","NH","NI","NJ","NK","NL","NN","PA","PB","PC","PD","PE","PF","PG","PH","PK"]
label=[['NA'],['NB'],['NC'],['ND'],['NE'],['NG'],['NH'],['NI'],['NJ'],['NK'],['NL'],['NN'],['PA'],['PB'],['PC'],['PD'],['PE'],['PF'],['PG'],['PH'],['PK']]

with open("stopwords.txt","r",encoding = 'utf8') as f:
    stopword_set=f.read().split("\n")


##將情緒標籤建立字典
##key:情緒標籤 value:之後鍵入的情緒相似度數值(目前為0)
dictwor={}
for i in range(len(itemize)):
    dictwor[itemize[i]]=0
```

```python
import jieba
from gensim.models import word2vec
from gensim import models
import redis
r = redis.Redis(host='localhost',port=6379,db=0,decode_responses=True)

songs={}
model = models.Word2Vec.load('word2vec.model')
songlabel={}
songlabelfile=open ("songlabelfile.txt","w+",encoding = 'utf8')
num=-1

for linkss in range(len(links)):
    num+=1
    print("完成"+str(num)+"筆")
    respin = requests.get(links[linkss])
    soupin = BeautifulSoup(respin.text, 'html5lib')
    aa=soupin.select("#fsZx3")
    bb=soupin.select("#fsZx2")
    cc=soupin.select("h1 > a[href]")
    
    for nn in range(len(aa)):
        lyric=str(aa[nn]).split('<dd class="fsZx3" id="fsZx3"><br/>')[1].split('[')[0].split("<ol><li> 感謝")[0].replace( '<br/>更多更詳盡歌詞 在 <a href="http://mojim.com">※ Mojim.com\u3000魔鏡歌詞網 </a>',"").replace("<br/>"," ").replace("\u3000","").replace("\n</dd>","").replace(" ","").replace("作詞","").replace("作曲","").replace("編曲","")
        title=str(bb[nn]).split('<dt class="fsZx2" id="fsZx2"> ')[1].split('</dt>')[0]
        singer=str(cc[0]).split('title="')[1].split(' 共收藏')[0]
        songs[title+singer]=lyric
    
        query = lyric
        tokens = jieba.lcut(query, HMM=True) 

        q_list = []
        for token in tokens:
             if token not in stopword_set:
                q_list.append(token)

        predword={} 
        for prednum in range(len(q_list)):
            predword[q_list[prednum]]="0"
        
        dictwor2={}
        for keyword in predword.keys():      
            for inword in predword[keyword]:    
                count={}
                for itemnum in range(-1,20):
                    a=str(itemize[itemnum+1])
                    if keyword in emotionless[a]: 
                        dictwor2[keyword]=[a,1000]
                        break
                    else:continue

        for keyword in predword.keys():      
            if keyword not in dictwor2.keys():    
                    count={}
                    for itemnum in range(-1,20):
                        a=str(itemize[itemnum+1])                        
                        wordcount=0
                        emsimsum=0
                        for datainn in range(len(emotionless[itemize[itemnum+1]])):
                            try:
                                emsimsum+=model.similarity(keyword,emotionless[itemize[itemnum+1]][datainn])
                                inwordcount+=1
                            except Exception as e:
                                repr(e)
                                continue
                        count[a]=emsimsum/(wordcount+1)
                        dictwor2[keyword]=count
        
        worddict={}

        for dictnum in dictwor2:
            if type(dictwor2[dictnum]) == list:
                worddict[dictnum]=dictwor2[dictnum][0]
            else:
                worddict[dictnum]=sorted(dictwor2[dictnum].items(),key=lambda item:item[1], reverse = 1 )[0][0]

        emolabelcount={}
        elist=list(worddict.values())

        for itemnum in range(-1,20):
            a=str(itemize[itemnum+1])  
            emolabelcount[a]=elist.count(a)

        songemotion=sorted(emolabelcount.items(),key=lambda item:item[1], reverse = 1 )[0][0]
        
        ##key:歌名 value:label
        
        songlabel[title]=songemotion
        songlabelfile.write(title+" "+singer+" "+lyric+" "+songemotion+"\n")
        r.rpush(songemotion,title) 
        print(title+"end")

songlabelfile.close()        
```

```python
sortout={}
for emott in itemize:
    esort=[]
    for kk in songlabel.keys():
        if songlabel[kk]==emott:
            esort.append(kk)
    sortout[emott]=esort

sortout
```
